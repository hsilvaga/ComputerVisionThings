```python
%pylab inline
import cv2
```

```python
img1 = imread('../images/small/image_00.jpg')
gray1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
```

```python
img2 = imread('../images/small/image_03.jpg')
gray2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
ORB = cv2.ORB_create(nfeatures = 2000)

kpts1, des1 = ORB.detectAndCompute(gray1, None)
kpts2, des2 = ORB.detectAndCompute(gray2, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING)
matches = bf.knnMatch(des1, des2, k = 2) #2 nearest matches
```

```python
#Use lowes ratio
matches_cleaned = []
for m,n in matches:
    if m.distance < .75 * n.distance:
        matches_cleaned.append(m)
```

```python
matches_cleaned = sorted(matches_cleaned, key = lambda x:x.distance)
```

```python
matching_keaypoints_1 = []
matching_keaypoints_2 = []

for match in matches_cleaned:
    matching_keaypoints_1.append(kpts1[match.queryIdx])
    matching_keaypoints_2.append(kpts2[match.trainIdx])
```
